```python
from backtesting import Backtest,Strategy
#import pandas_ta as ta
import pandas_datareader as dr
import pandas as pd
from backtesting.lib import crossover, cross
from backtesting.test import GOOG
```

```python
import talib as ta
def EMA_Backtesting(values, n):
    close = pd.Series(values)
    return ta.EMA(close, timeperiod=n)

class My9EMAStratergy(Strategy):
    def init(self):
        self.close = self.data.Close
        self.low = self.data.Low
        self.ema9 = self.I(EMA_Backtesting, self.data.Close, 9)  # Use self.I to calculate indicators
        self.ema21 = self.I(EMA_Backtesting, self.data.Close, 21) # Use self.I to calculate indicators

    def next(self):
        if (self.low[-1] < self.ema9[-1]) and (self.close[-1] > self.ema9[-1]):
            self.buy()
        elif (self.close[-1] < self.ema21[-1]):  # Access current bar's data with [-1]
            self.sell()
```

```python
start_date = ('2020-01-01')
end_date = ('2025-01-01')
#data = dr.DataReader("TSLA", "yahoo",start_date, end_date)
data = GOOG
bt = Backtest(data, My9EMAStratergy, cash=10000, commission=.002, exclusive_orders=True)
stat = bt.run()
```
